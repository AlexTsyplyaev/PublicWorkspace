```python
# Считывание, нормализация,подготовка данных
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
data = pd.read_csv('./ENB2012_data.csv',engine = 'python',encoding='ascii')
data.columns = [u'Relative Compactness',u'Surface Area',u'Wall Area',
u'Roof Area',
u'Overall Height',
u'Orientation',
u'Glazing Area',
u'Glazing Area Distribution',
u'Heating Load',
u'Cooling Load']
for i in range (0,len(data['Cooling Load'].values)):
    data['Cooling Load'].values[i]=int(data['Cooling Load'].values[i])
y= data['Cooling Load']
X = data.drop(('Heating Load'), axis = 1).drop(('Cooling Load'), axis = 1)
```

```python
# Разделяем данные для train и test выборок
from sklearn import tree
from sklearn import cross_validation
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 31)
```

```python
#Predefinitions
estimators=np.arange(1, 100)
def optimal_values(grid):
    return (1-grid.best_score_,grid.best_estimator_.n_estimators)
def print_optimal_values(grid):
    print(optimal_values(grid))
def plot_dep(grid):
    errors = 1 - np.array([cvres for cvres in grid.cv_results_['mean_test_score']])
    estim = np.array([cvres['n_estimators'] for cvres in grid.cv_results_['params']])
    plt.scatter(estim, errors)
```

```python
# Random Forests
model = ensemble.RandomForestRegressor()
grid = GridSearchCV(model, param_grid = {'n_estimators': estimators})
grid.fit(X, y)
print_optimal_values(grid)
```

```python
plot_dep(grid)
```

```python
cols,ers,est=[i for i in range(1, len(X.columns))],[],[]
for col in cols:
    model = ensemble.RandomForestRegressor()
    localpca = PCA(n_components=col)
    clocalpca = localpca.fit_transform(X)
    grid = GridSearchCV(model, param_grid = {'n_estimators': estimators})
    grid.fit(clocalpca, y)
    est.append(optimal_values(grid)[1])
    ers.append(optimal_values(grid)[0])
    print_optimal_values(grid)
```

```python
#Such 3d, so vision, much realistic
f = plt.figure()
splot = f.add_subplot(111, projection='3d')
splot.scatter(cols,ers,est)
splot.set_xlabel('Columns')
splot.set_ylabel('Error')
splot.set_zlabel('Num of Estimators')
plt.show()
```

```python
#Additional changes to dataset
data_ = (data - data.mean(axis = 0))/data.std(axis = 0)
for i in range (0,len(data_['Cooling Load'].values)):
    data_['Cooling Load'].values[i]=int(data_['Cooling Load'].values[i])
y= data_['Cooling Load']
X = data_.drop(('Heating Load'), axis = 1).drop(('Cooling Load'), axis = 1)
```

```python
# ExtraTreesClassifier
model = ensemble.ExtraTreesClassifier()
grid = GridSearchCV(model, param_grid = {'n_estimators': estimators})
grid.fit(X, y)
print_optimal_values(grid)
```

```python
plot_dep(grid)
```

```python
# GradientBoostingClassifier
model = ensemble.GradientBoostingClassifier()
grid = GridSearchCV(model, param_grid = {'n_estimators': estimators})
grid.fit(X, y)
print_optimal_values(grid)
```

```python
plot_dep(grid)
```

```python
estimators = np.arange(optimal_values(grid)[1]-5, optimal_values(grid)[1]+5, 2)
lr_range = np.arange(0.1, 3, 0.15)
deep_range = np.arange(1,20)
model = ensemble.GradientBoostingClassifier()
grid = GridSearchCV(model, param_grid = {'learning_rate': lr_range, 'n_estimators': estimators,'max_depth':deep_range})
grid.fit(X, y)
best_lr = grid.best_estimator_.learning_rate
best_depth = grid.best_estimator_.max_depth
print("for lr ",best_lr, "and depth ", best_depth)
print_optimal_values(grid)
```

```python
# AdaBoostClassifier
model = ensemble.AdaBoostClassifier()
grid = GridSearchCV(model, param_grid = {'n_estimators': estimators})
grid.fit(X, y)
print_optimal_values(grid)
```

```python
estimators = np.arange(optimal_values(grid)[1]-5, optimal_values(grid)[1]+5, 2)
lr_range = np.arange(0.1, 3, 0.15)
deep_range = np.arange(1,20)
for i in deep_range:
    model = ensemble.AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=i))
    grid = GridSearchCV(model, param_grid = {'learning_rate': lr_range, 'n_estimators': estimators})
    grid.fit(X, y)
    best_lr = grid.best_estimator_.learning_rate
    print("for lr ",best_lr, "and depth ", i)
    print_optimal_values(grid)
```
